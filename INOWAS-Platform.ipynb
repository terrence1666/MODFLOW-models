## 1. General

**For setting up the general properties on the platform:**
* The cell size can not be defined when a new model is created

```python
Image('pic/createmodel.png')
```

* The units are fixed and can not be changed

```python
Image('pic/units.png')
```

**For uploading a shape file to define the model area:**

* the mesh girds are always horizontal, which means the girds would not rotate when the projection is different from EPSG:3857 (The projection for Open Street Map). Rotation could be added in this part.

```python
Image('pic/grid.png')
```

**There is no option to define 'starting heads' and 'ibound' for models on the platform:**
* For starting heads, it has been defined as the same value as the top elevation.
* For ibound, all the cells are assumed to be active. However, sometimes constant boundary is not always equal to Time Variant Boundary (E.g. Using constant boundary to present seawater)

```python
Image('pic/ibound.png')
```
